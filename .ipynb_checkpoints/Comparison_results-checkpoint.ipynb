```python
#1) Import libraries
import gym
import math
import random
import matplotlib
import matplotlib.pyplot as plt
import torch
import numpy as np
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from collections import namedtuple, deque
import Environments
from collections import OrderedDict
```

```python
#2) Import and reshape reward lists
ddpg=np.load('load_model/parameters/ddpg4-02-reward-2000000.npy')
ppo=np.load('load_model/parameters/ppo-1-reward-001000.npy')
ppo_reshape=ppo.reshape(-1,10).mean(axis=1)
ddpg_reshape=ddpg[:1000000].reshape(-1,20).mean(axis=1)
print(len(ppo_reshape))
print(len(ddpg_reshape))
```

```python
#3) Plot comparison reward versus step
x_ppo=[i*2000 for i in range(len(ppo_reshape))]
x_ddpg=[i*40 for i in range(len(ddpg_reshape))]
#x_ddpg=[i for i in range(len(ddpg_reshape))]
plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
plt.title('Rewards over steps')
plt.xlabel("Steps")
plt.ylabel("Reward value")
plt.plot(x_ppo,ppo_reshape,color='cornflowerblue',label='ppo')
plt.plot(x_ddpg,ddpg_reshape,color='olivedrab',label='ddpg')
plt.grid(True,color='lightgrey')
plt.legend()
#plt.savefig("PDF_results/reward_comparison.pdf")
```
